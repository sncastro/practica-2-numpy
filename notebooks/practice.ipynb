```python
import numpy as np  # Importación de librerías
import matplotlib.pyplot as plt
```

### 1. **Generación de datos**

```python
dist_normal = np.random.normal(loc=50, scale=10, size=1000)  # Defino distribución normal
dist_uniforme = np.random.uniform(low=20, high=80, size=1000)  # Defino distribución uniforme
```

### 2. Análisis estadístico

```python
print("Estadísticos: Distribución Normal")

print(f"Media:               {np.mean(dist_normal):.4f}")
print(f"Mediana:             {np.median(dist_normal):.4f}")
print(f"Varianza:            {np.var(dist_normal):.4f}")
print(f"Desviación estándar: {np.std(dist_normal):.4f}")
print(f"Mínimo:              {np.min(dist_normal):.4f}")
print(f"Máximo:              {np.max(dist_normal):.4f}")
```

```python
print("Estadísticos: Distribución Uniforme")

print(f"Media:               {np.mean(dist_uniforme):.4f}")
print(f"Mediana:             {np.median(dist_uniforme):.4f}")
print(f"Varianza:            {np.var(dist_uniforme):.4f}")
print(f"Desviación estándar: {np.std(dist_uniforme):.4f}")
print(f"Mínimo:              {np.min(dist_uniforme):.4f}")
print(f"Máximo:              {np.max(dist_uniforme):.4f}")
```

### 3. Visualización

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))


axes[0].hist(dist_normal, bins=30, color='skyblue', edgecolor='black', alpha=0.7)  # Subplot izquierdo, distribución Normal
axes[0].axvline(np.mean(dist_normal), color='red', linestyle='--',
                linewidth=2, label=f'Media: {np.mean(dist_normal):.2f}')
axes[0].set_title('Distribución Normal (Gaussiana)', fontsize=14, fontweight='bold')
axes[0].set_xlabel('Valores', fontsize=12)
axes[0].set_ylabel('Frecuencia', fontsize=12)
axes[0].legend()
axes[0].grid(True, alpha=0.3)


axes[1].hist(dist_uniforme, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)  # Subplot derecho, distribución uniforme
axes[1].axvline(np.mean(dist_uniforme), color='red', linestyle='--',
                linewidth=2, label=f'Media: {np.mean(dist_uniforme):.2f}')
axes[1].set_title('Distribución Uniforme', fontsize=14, fontweight='bold')
axes[1].set_xlabel('Valores', fontsize=12)
axes[1].set_ylabel('Frecuencia', fontsize=12)
axes[1].legend()
axes[1].grid(True, alpha=0.3)

fig.suptitle('Análisis de Distribuciones Probabilísticas',  # Título general
             fontsize=16, fontweight='bold')

plt.tight_layout()
plt.show()
```

### 4. Reflexión

Se observan las siguientes diferencias entre las distribuciones:
- Distribución Normal:
    - Forma de campana simétrica centrada sobre la media.
    - La media y la mediana son prácticamente iguales, lo que refleja la distribución simétrica de los datos.
    - La desviación estándar que definí es moderada porque representa un 20% del valor de la media. Esto genera una muestra que varía razonablemente
- Distribución Uniforme:
    - Presenta una forma rectangular, esto se traduce a que la mayoría de los valores tienen igualdad de probabilidad.
    - Las frecuencias son constantes a lo largo del rango
    - Tiene una mayor desviación estándar
    - No se observa una moda clara en los datos

**Conclusión:**

Presentar los datos gráficamente ayuda a analizar de forma más sencilla distintos tipos de distribuciones.

La distribución normal muestra una concentración de valores alrededor de la media (con forma de campana), mientras que la distribución uniforme
presenta frecuencias constantes en todo su rango.
